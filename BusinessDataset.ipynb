```python
import pandas as pd

# Normalize and clean 'city' and 'categories' columns
business_df['city'] = business_df['city'].str.lower().str.strip()
business_df['categories'] = business_df['categories'].str.lower().str.strip()

# Filter for businesses located in Alberta, specifically in Edmonton
alberta_edmonton = business_df[
    (business_df['state'] == 'AB') &
    (business_df['city'].str.contains('edmonton', case=False))
]

# Define restaurant-related keywords
restaurant_keywords = [
    'restaurant', 'food', 'vegan', 'cuisine', 'bar', 'cafe',
    'diner', 'bistro', 'steakhouse', 'persian', 'healthy',
    'mexican', 'halal', 'indian', 'pizzeria', 'buffet',
    'brunch', 'takeout', 'fast food', 'bakery', 'grill',
    'bbq', 'catering', 'sandwich', 'seafood', 'sushi',
    'tavern', 'coffee', 'tea', 'pub', 'winery', 'dessert'
]

# Filter for restaurant-related businesses using restaurant keywords
restaurants_in_edmonton = alberta_edmonton[
    alberta_edmonton['categories'].notna() &
    alberta_edmonton['categories'].str.contains('|'.join(restaurant_keywords), case=False)
]

# Define cuisine keywords for specific categorization
cuisine_keywords = {
    'halal': ['halal', 'middle eastern', 'arabic'],
    'indian': ['indian', 'curry', 'tandoori', 'biryani'],
    'mexican': ['mexican', 'taco', 'burrito', 'quesadilla'],
    'vegan': ['vegan', 'vegetarian', 'plant-based', 'no meat'],
    'persian': ['persian', 'kebab', 'ghormeh sabzi'],
    'pizzeria': ['pizza', 'pizzeria', 'calzone'],
    'italian': ['italian', 'pasta', 'lasagna', 'risotto'],
    'japanese': ['japanese', 'sushi', 'ramen', 'tempura'],
    'chinese': ['chinese', 'noodles', 'dim sum', 'wontons'],
    'thai': ['thai', 'pad thai', 'curry', 'tom yum'],
    'french': ['french', 'crepes', 'bistro', 'ratatouille'],
    'korean': ['korean', 'kimchi', 'bulgogi'],
    'greek': ['greek', 'gyros', 'moussaka'],
    'cajun': ['cajun', 'creole', 'gumbo'],
    'turkish': ['turkish', 'kebab', 'baklava'],
    'kosher': ['kosher', 'koshert', 'matzo', 'challah'],
    'seafood': ['seafood', 'lobster', 'shrimp', 'oyster'],
    'lebanese': ['lebanese', 'hummus', 'falafel', 'tabbouleh'],
    'spanish': ['spanish', 'paella', 'tapas'],
    'vietnamese': ['vietnamese', 'pho', 'banh mi'],
    'brazilian': ['brazilian', 'churrasco', 'feijoada'],
    'ethiopian': ['ethiopian', 'injera', 'wat'],
    'american': ['american', 'burger', 'steak', 'bbq', 'fried chicken', 'hot dog'],
    'other': ['restaurant']  # Catch-all category
}

# Apply the function to create the 'cuisine' column
def extract_cuisine(categories):
    if pd.isna(categories):
        return 'other'
    for cuisine, keywords in cuisine_keywords.items():
        if any(keyword in categories for keyword in keywords):
            return cuisine
    return 'other'

restaurants_in_edmonton['cuisine'] = restaurants_in_edmonton['categories'].apply(extract_cuisine)

# Filter for open restaurants
open_restaurants = restaurants_in_edmonton[restaurants_in_edmonton['is_open'] == 1]

# Remove duplicates based on 'name' and 'address'
open_restaurants_sorted = open_restaurants.drop_duplicates(subset=['name', 'address'], keep='first')

# Drop rows with missing values in important columns (e.g., 'stars' and 'review_count')
open_restaurants_sorted.dropna(subset=['stars', 'review_count'], inplace=True)

# Drop unnecessary columns
columns_to_drop = [
    'address', 'city', 'state', 'postal_code',
    'categories', 'attributes', 'hours'  # Adjust as needed
]
open_restaurants_sorted.drop(columns=columns_to_drop, inplace=True, errors='ignore')

# Reset the index after cleaning
open_restaurants_sorted.reset_index(drop=True, inplace=True)

# Display sample cleaned data without 'categories' column
print("Sample cleaned data:")
display(open_restaurants_sorted[['name', 'cuisine', 'stars']].tail())

# Display the final shape of the cleaned DataFrame
print("\nCleaned DataFrame shape after preprocessing:", open_restaurants_sorted.shape)

# Check for remaining missing values
print("\nMissing values in cleaned DataFrame:")
print(open_restaurants_sorted.isnull().sum())
```
